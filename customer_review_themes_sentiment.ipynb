```python
def split_and_align(row):
    # Ensure the row values are scalar (convert arrays/lists to strings if necessary)
    theme_value = row['Theme'] if isinstance(row['Theme'], str) else str(row['Theme'])
    sentiment_value = row['Sentiment'] if isinstance(row['Sentiment'], str) else str(row['Sentiment'])

    # Handle missing or invalid values
    themes = theme_value.split(', ') if pd.notnull(theme_value) else []
    sentiments = sentiment_value.split(', ') if pd.notnull(sentiment_value) else []

    # Match themes with sentiments (ensure lengths match)
    if len(themes) == len(sentiments):
        return pd.DataFrame({'Theme': themes, 'Sentiment': sentiments})
    else:
        print(f"Mismatched lengths in row: {row}")
        return pd.DataFrame()
```

```python
# Apply the function to each row and concatenate the results
aligned_data = pd.concat(
    [split_and_align(row) for _, row in BSC_data.iterrows()],
    ignore_index=True
)

# Display the resulting DataFrame
print("Aligned DataFrame:")
print(aligned_data)
```

```python

import numpy as np
def clean_data(value):
    """
    Clean the input value by removing square brackets and extra quotes.
    """
    if isinstance(value, str):  # If the value is a string, clean it
        if pd.isnull(value):  # Handle NaN values
            return None
        return value.replace("[", "").replace("]", "").replace("'", "").strip()
    elif isinstance(value, (list, np.ndarray)):  # If the value is a list or array
        return ', '.join(str(item) for item in value)  # Convert list/array to a comma-separated string
    elif pd.isnull(value):  # Handle NaN values for non-string types
        return None
    else:
        return str(value).strip()  # For other types, just return as string


def split_and_align(row):
    """
    Split the 'Theme' and 'Sentiment' columns, align them, and return a DataFrame.
    """
    # Clean the 'Theme' and 'Sentiment' values
    theme_value = clean_data(row['Theme'])
    sentiment_value = clean_data(row['Sentiment'])

    # Split the strings by ', ' if not null
    themes = theme_value.split(", ") if theme_value else []
    sentiments = sentiment_value.split(", ") if sentiment_value else []

    # Match themes with sentiments (ensure lengths match)
    if len(themes) == len(sentiments):
        return pd.DataFrame({'Theme': themes, 'Sentiment': sentiments})
    else:
        print(f"Mismatched lengths in row: {row}")
        return pd.DataFrame()


# Apply the function to each row and concatenate the results
aligned_data = pd.concat(
    [split_and_align(row) for _, row in BSC_data.iterrows()],
    ignore_index=True
)

# Display the resulting DataFrame
print("Aligned DataFrame:")
print(aligned_data)

```

```python
# Count occurrences of each theme per sentiment and sort in descending order
theme_sentiment_counts = aligned_data.groupby(['Theme', 'Sentiment']).size().reset_index(name='Count')

# Sort by 'Count' in descending order
theme_sentiment_counts_sorted = theme_sentiment_counts.sort_values(by='Count', ascending=False)

# Display the sorted result
print(theme_sentiment_counts_sorted)
```

```python
from langchain_community.llms import Ollama
import pandas as pd
import json

# Define the prompt
prompt = '''
You are a professional hair specialist with expertise in hair care, styling, and product recommendations. 
Your role is to provide personalized advice, answer customer queries about hair issues, and recommend 
suitable products based on their hair type, concerns, and goals. Please respond in a knowledgeable, empathetic, 
and customer-friendly tone.

Give the JSON output only and nothing else, which is a list of themes and sentiments in the customer review.
Example JSON:
{
  "themes": [
    "product effectiveness",
    "packaging"
  ],
  "sentiments": [
    {
      "theme": "product effectiveness",
      "sentiment": "positive",
      "strength": 5,
      "reasons": ["immediately felt and saw the difference it made on my hair"]
    },
    {
      "theme": "packaging",
      "sentiment": "neutral",
      "strength": 2,
      "reasons": ["wish the bottles were a little bigger"]
    }
  ]
}
'''

# Initialize the Ollama model
ollama_model = Ollama(model="llama3.2")
count=1
# Function to analyze reviews and extract themes and sentiments
def analyze_review(review, prompt):
    global count
    print(count)
    count=count+1
    input_text = f"Prompt: {prompt}\nReview: {review}\nOutput:"
    response = ollama_model.invoke(input_text)
    try:
        # Parse the JSON response
        response_json = json.loads(response)
        themes = ", ".join(response_json.get("themes", []))
        sentiments = ", ".join([item["sentiment"] for item in response_json.get("sentiments", [])])
        return themes, sentiments
    except (json.JSONDecodeError, KeyError):
        return None, None  # Return None if parsing fails

# Filter reviews where PRODUCT_ROLL_UP_NAME is 'Broad Spectrum Chelating'
# and REVIEW_REVIEWTEXT is valid
filtered_data = data[
    (data['PRODUCT_ROLL_UP_NAME'] == 'Brow Bond') & 
    (data['REVIEW_REVIEWTEXT'] != "Sorry, I can't assist with that.")
]

# Apply the function to extract themes and sentiments
filtered_data[['Theme', 'Sentiment']] = filtered_data[:20]['REVIEW_REVIEWTEXT'].apply(
    lambda review: pd.Series(analyze_review(review, prompt))
)

# Update the original DataFrame with the new columns for matching rows
data.loc[filtered_data.index, ['Theme', 'Sentiment']] = filtered_data[['Theme', 'Sentiment']]

# Save the updated DataFrame to a CSV file
data.to_csv('output_with_themes_and_sentiments1.csv', index=False)

# Print the resulting DataFrame with themes and sentiments
print(data)
```

```python
from langchain_community.llms import Ollama
import pandas as pd
import json

# Define the prompt
prompt = '''
You are a professional hair specialist with expertise in hair care, styling, and product recommendations. 
Your role is to provide personalized advice, answer customer queries about hair issues, and recommend 
suitable products based on their hair type, concerns, and goals. Please respond in a knowledgeable, empathetic, 
and customer-friendly tone.

Give the JSON output only and nothing else, which is a list of themes and sentiments in the customer review.
Example JSON:
{
  "themes": [
    "product effectiveness",
    "packaging"
  ],
  "sentiments": [
    {
      "theme": "product effectiveness",
      "sentiment": "positive",
      "strength": 5,
      "reasons": ["immediately felt and saw the difference it made on my hair"]
    },
    {
      "theme": "packaging",
      "sentiment": "neutral",
      "strength": 2,
      "reasons": ["wish the bottles were a little bigger"]
    }
  ]
}
'''

# Initialize the Ollama model
ollama_model = Ollama(model="llama3.2")

# Function to analyze reviews and extract themes and sentiments
def analyze_review(review, prompt):
    input_text = f"Prompt: {prompt}\nReview: {review}\nOutput:"
    response = ollama_model.invoke(input_text)
    try:
        # Parse the JSON response
        response_json = json.loads(response)
        sentiments = response_json.get("sentiments", [])
        # Return a list of theme-sentiment pairs
        return [
            {
                "theme": item["theme"],
                "sentiment": item["sentiment"],
                "strength": item.get("strength"),
                "reasons": item.get("reasons", [])
            }
            for item in sentiments
        ]
    except (json.JSONDecodeError, KeyError):
        return []  # Return an empty list if parsing fails

# Filter reviews where PRODUCT_ROLL_UP_NAME is 'Broad Spectrum Chelating'
# and REVIEW_REVIEWTEXT is valid
filtered_data = data[
    (data['PRODUCT_ROLL_UP_NAME'] == 'Broad Spectrum Chelating') & 
    (data['REVIEW_REVIEWTEXT'] != "Sorry, I can't assist with that.")
]

# Analyze reviews and expand the themes and sentiments into separate rows
expanded_results = []
for idx, review in filtered_data['REVIEW_REVIEWTEXT'].items():
    review_analysis = analyze_review(review, prompt)
    for result in review_analysis:
        expanded_results.append({
            "Index": idx,
            "Review": review,
            "Theme": result["theme"],
            "Sentiment": result["sentiment"],
            "Strength": result["strength"],
            "Reasons": ", ".join(result["reasons"])
        })

# Create a DataFrame from the expanded results
expanded_df = pd.DataFrame(expanded_results)

# Merge expanded results with the original data
final_data = pd.merge(
    data, expanded_df, left_index=True, right_on="Index", how="left"
)

# Save the updated DataFrame to a CSV file
final_data.to_csv('output_with_themes_and_sentiments2.csv', index=False)

# Print the resulting DataFrame
print(final_data)

```
